```python
import random

##list of all groups/ranks, with respective values
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}


##class Card used to setup/group attributes of card
class Card():
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]

    def __str__(self):
        return f"You have here {self.rank} of {self.suit}, where value is {self.value}"


#class Deck used to create a deck of cards, using Card class. it also allows to shuffle  the deck and pickup one card from whole deck
class Deck:
    def __init__(self):
        self.full_deck = []
        for suit in suits:
            for rank in ranks:
                self.full_deck.append(Card(suit, rank))

    
    def mix_cards(self):
        random.shuffle(self.full_deck)

    def deal_one(self):
        return self.full_deck.pop()

    def __str__(self):
        return self.full_deck


#class Player used to assign name and card set to a person. allows to add/remove cards from his hand and return string describing his setup
class Player:
    def __init__(self, name):
        self.name = name
        self.card_set = []

    def remove_card(self):
        return self.card_set.pop(0)

    def add_card(self, new_card):
        if type(new_card) == type([]):
            self.card_set.extend(new_card)
        else:
            self.card_set.append(new_card)

    def __str__(self):
        return f"{self.name} set of cards is {self.card_set}"


#setting up a new deck and mixing cards inside
set = Deck()

set.mix_cards()


#dividing deck into 2 groups, for each player
set1 = set.full_deck[0:26]
set2 = set.full_deck[26:52]

#setting up a player and his cards
player1 = Player('Januszex')
player1.add_card(set1)

player2 = Player('Marcinex')
player2.add_card(set2)

print(type(set.full_deck[2]))

## war game logic 



game_on = True


round_count = 0
while game_on:
    
    if len(player1.card_set) == 0:
        print("player 2 won nara")
        game_on = False
        break


    if len(player2.card_set) == 0:
        print("player 1 won nara")
        game_on = False
        break

    round_count += 1
    print(f'round {round_count}. FIGHT!')
    print(f'player 1 has {len(player1.card_set)}')
    print(f'player 2 has {len(player2.card_set)}')

    played_cards_1 = []
    played_cards_2 = []
    played_cards_1.append(player1.remove_card())
    played_cards_2.append(player2.remove_card())
    war = True


    while war:
        if played_cards_1[-1].value > played_cards_2[-1].value:
            print('player1 won round')
            player1.add_card(played_cards_2)
            player1.add_card(played_cards_1)
            war = False
        elif played_cards_1[-1].value < played_cards_2[-1].value:
            print('player2 won round')
            player2.add_card(played_cards_1)
            player2.add_card(played_cards_2)
            war = False
        else:
            
            if len(player1.card_set) < 5:
                print("we have a war, but player1 does not have enough cards to declare war, so spierdalaj. PLAYER 2 WINS")
                war = False
                game_on = False
                break
            
            elif len(player2.card_set) < 5:
                print("we have a war, but player2 does not have enough cards to declare war, so spierdalaj. PLAYER 1 WINS")
                war = False
                game_on = False
                break

            else:
                print("we have war my friend")
                for i in range(5):
                    played_cards_1.append(player1.remove_card())
                    played_cards_2.append(player2.remove_card())

```
