```python
#make selection for player 1 - either X or O
#randomello, who starts the game
#draw board to play

#make a default dictionary for each position in board, so initially every point has ' ' assigned
#ask for 1st player move between numbered board (1-9)
#update dictionary of positions
#update board screen, update 'used moves', update 'move counter', print board
#update player positions list
#ask for 2nd move (show reduced list of available moves)
#update board
#after 5th move in 'move counter' - start checking, if 'used moves' match with any of winning configurations


from IPython.display import clear_output

winning_configurations = [{1, 2, 3}, {4, 5, 6}, {7, 8, 9}, {1, 5, 9}, {3, 5, 7}, {1, 4, 7}, {2, 5, 8}, {3, 6, 9}]

def intro():
    print('welcome in the best kółko krzyzyk game !!!! XD')
    player1 = ''
    player2 = ''
    while player1 != 'x' and player1 != 'o':
        #clear_output()
        player1 = input('Gimme x or o for player1')
    if player1 == 'x':
        player2 = 'o'
    else:
        player2 = 'x'
    print(f'Player 1 will be {player1}, and player 2 will be {player2}')

    return player1, player2

positions = {'1': ' ', '2': ' ', '3': ' ', '4': ' ', '5': ' ', '6': ' ', '7': ' ', '8': ' ', '9': ' '} 

def print_board(positions):
    print('─────────────────────────')
    print('│       │       │       │')
    print(f'│   {positions["1"]}   │   {positions["2"]}   │   {positions["3"]}   │')
    print('│       │       │       │')
    print('─────────────────────────')
    print('│       │       │       │')
    print(f'│   {positions["4"]}   │   {positions["5"]}   │   {positions["6"]}   │')
    print('│       │       │       │')
    print('─────────────────────────')
    print('│       │       │       │')
    print(f'│   {positions["7"]}   │   {positions["8"]}   │   {positions["9"]}   │')
    print('│       │       │       │')
    print('─────────────────────────')



def game(player1, player2):
    winner_found = False
    avl_moves = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    p1_positions = []
    p2_positions = []
    total_moves = 0
    while winner_found == False and total_moves < 9:
        move_player_1 = input(f'Give position on board you take for {player1}. You can choose out of {avl_moves}')
        clear_output()
        total_moves += 1
        positions[move_player_1] = player1
        p1_positions.append(int(move_player_1))
        print(f'P1 positions are {p1_positions}')
        print(f'P2 positions are {p2_positions}')
        avl_moves.remove(int(move_player_1))
        print_board(positions)

        if total_moves >= 5 and winner_found == False:
            winner_found = check_if_win(set(p1_positions), 'player 1')

        move_player_2 = input(f'Give position on board you take for {player2}. You can choose out of {avl_moves}')
        total_moves += 1
        print(f'Ilosc ruchow to {total_moves}')
        positions[move_player_2] = player2
        p2_positions.append(int(move_player_2))
        #print(f'P2 positions are {p2_positions}')
        avl_moves.remove(int(move_player_2))
        #clear_output()
        print(f'P1 positions are {p1_positions}')
        print(f'P2 positions are {p2_positions}')
        print_board(positions)
        print(avl_moves)
        
        if total_moves >= 5 and winner_found == False:
            winner_found = check_if_win(set(p2_positions), 'player 2')
    print('nikt kurwa nie wygral')
    break
        


def check_if_win(pos_to_check, player):
    for number in range(8):
        if len(winning_configurations[number].intersection(pos_to_check)) == 3:
            print(f'wygrales kurwa {player}')
            return True
    return False



player1, player2 = intro()
print_board(positions)
game(player1, player2)
```
